```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path = '/content/insurance (1).csv'
data = pd.read_csv(file_path)
```

```python
age_vector = data['age'].values
smoker_factor = pd.Categorical(data['smoker'])
data_list = [data['age'], data['smoker'], data['expenses']]
matrix = np.array([data['age'], data['expenses']])
new_data_frame = pd.DataFrame({
    'Age': data['age'],
    'Smoking_Status': data['smoker'],
    'Expenses': data['expenses']
})
print(new_data_frame.head())
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['age'], data['smoker'].apply(lambda x: 1 if x == 'yes' else 0), c='blue', alpha=0.6, label='Smoking Status')
plt.title('Scatter Plot: Age vs Smoking Status')
plt.xlabel('Age')
plt.ylabel('Smoking (1 = Yes, 0 = No)')
plt.yticks([0, 1], ['No', 'Yes'])
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```

```python
filtered_data = data[(data['age'] >= 20) & (data['age'] <= 40)]
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(filtered_data['age'], filtered_data['smoker'].apply(lambda x: 1 if x == 'yes' else 0), c='green', alpha=0.6, label='Smoking Status')
plt.title('Scatter Plot: Age (20-40) vs Smoking Status')
plt.xlabel('Age')
plt.ylabel('Smoking (1 = Yes, 0 = No)')
plt.yticks([0, 1], ['No', 'Yes'])
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```

```python
smoker_expenses = data[data['smoker'] == 'yes']['expenses']
non_smoker_expenses = data[data['smoker'] == 'no']['expenses']
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot([smoker_expenses, non_smoker_expenses], labels=['Smokers', 'Non-Smokers'], patch_artist=True)
plt.title('Expenses: Smokers vs Non-Smokers')
plt.ylabel('Expenses')
plt.grid(alpha=0.3)
plt.show()
```

```python
smoker_gender_counts = data[data['smoker'] == 'yes']['sex'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
smoker_gender_counts.plot(kind='bar', color=['orange', 'blue'])
plt.title('Number of Smokers by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Smokers')
plt.xticks(rotation=0)
plt.grid(alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['age'], data['expenses'], c='red', alpha=0.6)
plt.title('Scatter Plot: Age vs Expenses')
plt.xlabel('Age')
plt.ylabel('Expenses')
plt.grid(alpha=0.3)
plt.show()
```
